SECTION A - Interview Theory


Q1. A Series stores data in a single labeled column, whereas a Dataframe organizes data into multiple rows and columns like a excel or spreadsheet. Both support operations such as .mean() and .sum() to calculate values.

Q2. "inplace=True" modifies the original object instead of creating a new one. Pandas now recommends avoiding it because it doesnt always improve performance and can make debugging or method chaining harder.
For example, i would avoid df.dropna(inplace=True) when i want to preserve the original dataset for comparison.

Q3. .fillna() keeps the data by replacing missing values with something meaningful, like the mean or a default value. .dropna() removes rows or columns containing missing values. I would use .fillna() for missing customer ages, but .dropna() if only a few incomplete survey responses exist and will not affect the analysis.

Q4. With errors='coerce', invalid values are automatically converted to NaN instead of stopping the program with an error. The default version raises an exception whenever it finds non-numeric data. For real-world datasets containing values like "N/A" or "unknown", I would use errors='coerce' because it allows cleaning without interrupting the workflow.

Q5. '.loc' selects data using row labels, while '.iloc' selects by numerical positions. For example, df.loc['A'] retrieves the row labeled A, whereas df.iloc[0] retrieves the first row regardless of its label. If the index starts at 101, df.loc[0] searches for the label 0, but df.iloc[0] still returns the row with index 101.

Q6. '.map()' is ideal for converting values using a dictionary or a simple function. '.apply()' is more flexible and runs custom logic on every element. '.replace()' directly swaps specific values without applying any computation. I use .map() for category mapping, .apply() for calculations, and .replace() for correcting incorrect values.

Q7. This warning appears when pandas is not sure whether you are modifying the original Dataframe or just a temporary copy.
For example: df[df["Marks"] > 50]["Result"] = "Pass" to fix this use '.loc' instead
df.loc[df["Marks"] > 50, "Result"] = "Pass"
this will clearly updates the intended rows.

Q8. 'pd.cut()' creates groups using fixed value ranges, while 'pd.qcut()' divides the data so each group contains nearly the same number of records. Since income data is usually unevenly distributed, I would choose pd.qcut() because it creates balanced Low, Mid, and High income groups, making comparisons more meaningful.

Section C – Applied on Indian Startups Dataset

Loading data set

```python
import pandas as pd
import numpy as np
df = pd.read_csv("indian_startups_fundingDADS.csv")
df
```

```python
#Q17

# (a)
print("Rows, Columns:", df.shape)

# (b)
print("\nMissing values:")
print(df.isnull().sum())

# (c)
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
#Cleaning

# Rename columns
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(" ", "_")
)

# Clean text columns
df["startup_name"] = df["startup_name"].str.strip().str.title()
df["city"] = df["city"].str.strip().str.title()

# Standardize city names
city_map = {
    "Bangalore":"Bengaluru",
    "Blr":"Bengaluru",
    "Bombay":"Mumbai",
    "Gurgaon":"Gurugram",
    "Delhi":"New Delhi",
    "Calcutta":"Kolkata",
    "Madras":"Chennai"
}

df["city"] = df["city"].replace(city_map)

# Standardize Industry
industry_map = {
    "Ai Ml":"AI/ML",
    "Aiml":"AI/ML",
    "Ai":"AI/ML",
    "Health Tech":"HealthTech",
    "Healthtech":"HealthTech",
    "Fin Tech":"Fintech",
    "Fintech":"Fintech",
    "Ed Tech":"Edtech",
    "Edtech":"Edtech",
    "E Commerce":"E-Commerce",
    "Ecommerce":"E-Commerce"
}

df["industry"] = df["industry"].str.title().replace(industry_map)

# Standardize Funding Stage
stage_map = {
    "Seed Round":"Seed",
    "Seed Funding":"Seed",
    "Series A":"Series A",
    "Series B":"Series B",
    "Series C":"Series C",
    "Pre Seed":"Pre-Seed",
    "Angel":"Angel"
}

df["funding_stage"] = df["funding_stage"].str.title().replace(stage_map)

# Convert Amount

df["amount"] = (
    df["amount"]
      .astype(str)
      .str.replace("₹","",regex=False)
      .str.replace("INR","",regex=False)
      .str.replace("Cr","",regex=False)
      .str.replace("Crore","",regex=False)
      .str.replace(",","",regex=False)
      .replace("Undisclosed",np.nan)
)

df["amount"] = pd.to_numeric(df["amount"],errors="coerce")

# Funding Date

df["funding_date"] = pd.to_datetime(
    df["funding_date"],
    errors="coerce",
    dayfirst=True
)

# Founded Year
df["founded_year"] = pd.to_numeric(df["founded_year"],errors="coerce")

# Remove duplicates
df = df.drop_duplicates()
```
