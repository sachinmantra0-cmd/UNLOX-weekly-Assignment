```python
import pandas as pd

df = pd.read_csv('indian_salaries.csv')
df
```

```python
#Two types of stats descriptive and inferential stats

avg_salary = df['salary_lpa'].mean()
median_sal = df['salary_lpa'].median()
print(avg_salary)
print(median_sal)
#
```

```python
# plot
import matplotlib.pyplot as plt
df['salary_lpa'].plot.hist(bins=20)
plt.axvline(df.salary_lpa.mean(), color='red' , label='Mean')
plt.axvline(df.salary_lpa.median(), color='green' , label='Median')
plt.legend()
plt.show()
#
```

```python
# most common city

df['city'].mode()
```

```python
df['city'].value_counts()
```

```python
# range of salary

lowest_sal = df['salary_lpa'].min()
highest_sal = df['salary_lpa'].max()
print(lowest_sal)
print(highest_sal)
#

sal_range = highest_sal - lowest_sal
print(sal_range)
#
```

```python
# find the variance

sal_var = df['salary_lpa'].var()
exp_var = df['experience_years'].var()

std_dev = df['salary_lpa'].std()
print(std_dev)
print(exp_var)
print(sal_var)
#
```

```python
# q1(25%)
# q2(50%)
# q3(75%)
# q4(100%)

# IQR = inter-quantile range = q3-q1

q1 = df['salary_lpa'].quantile(0.25)
q2 = df['salary_lpa'].quantile(0.50)
q3 = df['salary_lpa'].quantile(0.75)
p90 = df['salary_lpa'].quantile(0.90)
print(p90)
print(q1)
print(q2)
print(q3)

iqr = q3-q1
print(iqr)

```

```python
 lower_bound = q1 - 1.5 * iqr
 upper_bound = q3 + 1.5 * iqr

 print(lower_bound)
 print(upper_bound)
#
```
